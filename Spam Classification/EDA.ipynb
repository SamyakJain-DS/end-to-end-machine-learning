```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('datasets/data_v2.csv')
```

```python
data.shape
```

```python
data['number_of_words'] = data['mail'].apply(lambda x: len(x.split()))
data['number_of_chars'] = data['mail'].apply(lambda x: len(x))
```

```python
data[data['number_of_words'] < 10]
```

```python
data.iloc[176,0]
```

```python
data[(data['number_of_words'] < 10) & (data['number_of_chars'] > 90)]['mail'].values
```

```python
drop_index = data[(data['number_of_words'] < 10) & (data['number_of_chars'] > 90)].index
```

```python
data = data.drop(index = drop_index).reset_index(drop=True)
```

```python
drop_index2 = data[(data['number_of_words'] < 25) & (data['number_of_chars'] > 200)].index
```

```python
data = data.drop(index = drop_index2).reset_index(drop=True)
```

```python
data
```

```python
data.groupby('label')['number_of_words'].mean().plot.bar()
```

```python
data.groupby('label')['number_of_chars'].mean().plot.bar()
```

```python
data['label'].value_counts().plot.bar()
```
